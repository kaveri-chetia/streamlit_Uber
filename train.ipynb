```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
import pickle
```

```python
df_clean =  pd.read_csv("./data/cleaned_data.csv")
df_clean
```

```python
np.random.seed(42)
data = {
    'trip_distance_km': np.random.rand(100),
    'Year': np.random.rand(100),
    'pickup_longitude': np.random.rand(100),
    'pickup_latitude': np.random.rand(100),
    'dropoff_longitude': np.random.rand(100),
    'dropoff_latitude': np.random.rand(100),
    'passenger_count': np.random.rand(100),
    'Month_sin': np.random.rand(100),
    'Month_cos': np.random.rand(100),
    'Hour_sin': np.random.rand(100),
    'Hour_cos': np.random.rand(100),
    'day_sin': np.random.rand(100),
    'day_cos': np.random.rand(100),
    'fare_amount': np.random.randint(0, 2, 100)
}
# df = pd.DataFrame(data)
```

```python
y = df_clean['fare_amount']
X = df_clean.drop(columns=['fare_amount', 'pickup_datetime','counter','Day of Week','Day of Week_num','Hour','Month'])
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X)


# Save the scaler to a file
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```

```python
gb_model = GradientBoostingRegressor(max_depth=10, n_estimators=100, random_state=42)
gb_model.fit(X_train_scaled, y)
```

```python
with open('gradientboost_model.pkl', 'wb') as file:
    pickle.dump(gb_model, file)
print("Model saved successfully!")
```

```python
with open('gradientboost_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
print("Model loaded successfully!")
```

```python
original_prediction = gb_model.predict(X)
loaded_prediction = loaded_model.predict(X)
print("Original model accuracy:", gb_model.score(X_train_scaled, y))
print("Loaded model accuracy:", loaded_model.score(X_train_scaled, y))
if np.array_equal(original_prediction, loaded_prediction):
    print("The loaded model produces the same predictions as the original model.")
else:
    print("Something went wrong. The predictions don't match.")
```
